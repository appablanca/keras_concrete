```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

data_a = pd.read_csv('concrete_data.csv')
```

Part A

```python
model_a = Sequential()
ncolls_a = data_a.shape[1] - 1
model_a.add(Dense(10, activation='relu', input_shape=(ncolls_a,)))
model_a.add(Dense(1))
model_a.compile(optimizer='adam', loss='mean_squared_error')

X_a = data_a.drop('Strength', axis=1)
y_a = data_a['Strength']
X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(X_a, y_a, test_size=0.3, random_state=42)

model_a.fit(X_train_a, y_train_a, epochs=50)
```

```python
lossvalues_a = []
for i in range(50):
    model_a.fit(X_train_a, y_train_a, epochs=50, verbose=0)
    lossvalues_a.append(model_a.evaluate(X_test_a, y_test_a))
```

```python
mean_a = np.mean(lossvalues_a)
print("Mean of lossvalues",mean_a)
std_a = np.std(lossvalues_a)
print("Standard deviation of lossvalues",std_a)
```

Part B

```python
X_train_b = X_train_a
y_train_b = y_train_a
X_test_b = X_test_a
y_test_b = y_test_a
```

```python
X_train_b = (X_train_b - X_train_b.mean()) / X_train_b.std()
X_test_b = (X_test_b - X_test_b.mean()) / X_test_b.std()
y_train_b = (y_train_b - y_train_b.mean()) / y_train_b.std()
y_test_b = (y_test_b - y_test_b.mean()) / y_test_b.std()
```

```python
lossvalues_b = []
for i in range(50):
    model_a.fit(X_train_b, y_train_b, epochs=50, verbose=0)
    lossvalues_b.append(model_a.evaluate(X_test_b, y_test_b))

mean_b = np.mean(lossvalues_b)
print("Mean of lossvalues",mean_b)
std_b = np.std(lossvalues_b)
print("Standard deviation of lossvalues",std_b)
```

normalizing the train and test data resulted in smaller loss values and better accuracy.
